```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
```

```python
(x_train,y_train),(x_test,y_test)=fashion_mnist.load_data()
```

```python
from tensorflow.keras.models import Sequential #model
from tensorflow.keras.layers import Dense,Flatten #layers->Dense->neurons
```

```python
model=Sequential()
```

```python
model.add(Flatten(input_shape=[28,28])) #input layer (28,28)
model.add(Dense(200,activation="relu"))
model.add(Dense(100,activation="relu"))
model.add(Dense(10,activation="softmax"))# softmax-> probability
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"]) #optimizer->speed up,loss->error
```

```python
model.fit(x_train,y_train,epochs=20) #epochs->relearning
```
